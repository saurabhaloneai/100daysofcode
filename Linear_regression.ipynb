## Simple linear regression model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

x_train = np.array([1.,2.])
y_train = np.array([300,500])
```

```python
plt.scatter(x_train,y_train,marker = 'x', c = 'r')
plt.title('House price predicting')
plt.xlabel('Price of houses')
plt.ylabel('Size of house in 1000 ft')

plt.show()
```

```python
w = 200
b = 100
```

```python
#Liner regression model 

def compute_price(x,w,b):
    m = len(x_train)
    i  = 0 
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
#Passing the values in model 

t_fwb  = compute_price(x_train,w,b)

#ploting our model prediction 

plt.plot(x_train,t_fwb , c = 'b', label = 'our prediction')

#ploting the data points 

plt.scatter(x_train, y_train , marker = 'x',c = 'r' , label = 'actual values')

plt.title("House prediction")
plt.xlabel('Size of house')
plt.ylabel('Proice of house')
plt.legend()
plt.show()
```

```python
#making prediction

w = 200
b = 100

x_i = 1.6

f_wb = w * x_i + b

print(f'the price of house: {f_wb}')

```

## Conclusion

##This is the simple linear regression model that take input x and y (target)  and predict ^y.

##Once we determined the values of w , b (parameters) we can use this to predict any house price from noval data.
