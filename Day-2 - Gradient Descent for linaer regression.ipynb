```python
## DATA of house x = size of in feet and y = price of house

import numpy as np

x= np.array([1.0,2.0])
y = np.array([300.0,500.0])
```

## Gradient Descent

```python

def compute_gradient(x, y, w, b, learning_rate): 
    dldw = 0.0 
    dldb = 0.0 
    N = x.shape[0]
    # loss = (y-(wx+b)))**2
    for xi, yi in zip(x,y): 
       dldw += -2*xi*(yi-(w*xi+b))
       dldb += -2*(yi-(w*xi+b))
    
    # Make an update to the w parameter 
    w = w - learning_rate*(1/N)*dldw
    b = b - learning_rate*(1/N)*dldb
    return w, b 
```
